```python
#Data visualizing and cleaning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
#Traning Testing and Spliting Data
from sklearn.model_selection import train_test_split
#for label encoding 
from sklearn.preprocessing import LabelEncoder
# Scale the data to be between -1 and 1
from sklearn.preprocessing import StandardScaler
#for SVM
from sklearn.svm import SVC
from sklearn import metrics
#Using Cross Validation
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('../Dataset/voice.csv')
df.head()
```

```python
df.corr()
```

```python
plt.figure(figsize=(20,20))
data_corr = df.corr()
sb.heatmap(data_corr,annot=True)
```

```python
print("Total number of labels: {}".format(df.shape[0]))
print("Number of male: {}".format(df[df.label == 'male'].shape[0]))
print("Number of female: {}".format(df[df.label == 'female'].shape[0]))
```

```python
df.isnull().sum()
```

```python
X = df.drop('label', axis=1)
y = df['label']
```

```python
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
y
```

```python
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
svc=SVC(kernel='poly') #Default hyperparameters
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
print('Accuracy Score\'s are:')
acc=[]
gamma=['scale','auto']
for i in gamma:
    for j in range(10):
        svc=SVC(kernel='poly',degree=j,gamma=i ) 
        svc.fit(X_train,y_train)
        y_pred=svc.predict(X_test)
        print('For degree ',j,'gamma as ',i,' =',metrics.accuracy_score(y_test,y_pred))
```

```python
#for gamma as scale
C_range=[0.1,1,10]
acc_score=[]
for c in C_range:
    svc = SVC(kernel='poly', C=c,degree=1,gamma='scale')
    scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score)  
```

```python
C_range=[0.1,1,10]
plt.figure(figsize=(20,20))


plt.plot(C_range,acc_score)
plt.xlabel('Value of C for SVC ')
plt.xticks(np.arange(0.1,10,1.5))
plt.ylabel('Cross-Validated Accuracy')
```

```python
print('Accuracy Score\'s are:')

svc=SVC(kernel='poly',gamma='scale',degree=1,C=1.5  ) 
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```

```python
#for gamma as auto
C_range=[0.1,1,10]
acc_score=[]
for c in C_range:
    svc = SVC(kernel='poly', C=c,degree=1,gamma='auto')
    scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score) 
```

```python
C_range=[0.1,1,10]
plt.figure(figsize=(20,20))


plt.plot(C_range,acc_score)
plt.xlabel('Value of C for SVC ')
plt.xticks(np.arange(0.1,10,1.5))
plt.ylabel('Cross-Validated Accuracy')
```
